# **:)  Data PreProcessing with the Vehicles(four wheel drive) Dataset**


```python
import pandas as pd
import numpy as np
headers_data = ["symboling","normalized-losses","make","fuel-type","aspiration", "num-of-doors","body-style",
         "drive-wheels","engine-location","wheel-base", "length","width","height","curb-weight","engine-type",
         "num-of-cylinders", "engine-size","fuel-system","bore","stroke","compression-ratio","horsepower",
         "peak-rpm","city-mpg","highway-mpg","price"]
```

```python
df = pd.read_csv('/dataset_1.data', names=headers_data)
df.head(10)
```

```python
df.dtypes
```

```python
df.describe(include = "all")
```

```python
df.replace("?", np.nan, inplace = True)
df.head(10)
```

```python
df.info()
```

```python
missing_data = df.isnull()

for columns in headers_data:
  print(columns)
  print(missing_data[columns].value_counts())
  print(" ")
```

# **Converting all the Object data type to the interger or float**

```python
avg_nor_loss = df['normalized-losses'].astype("float").mean()
avg_nor_loss
```

```python
df['normalized-losses'].replace(np.nan, avg_nor_loss, inplace=True)
```

```python
avg_stroke = df['stroke'].astype("float").mean()
df['stroke'].replace(np.nan, avg_stroke, inplace=True)
```

```python
avg_bore = df['bore'].astype("float").mean()
df['bore'].replace(np.nan, avg_bore, inplace=True)
```

```python
avg_horsepower = df['horsepower'].astype("float").mean()
df['horsepower'].replace(np.nan, avg_horsepower, inplace=True)
```

```python
df['num-of-doors'].value_counts()
df['body-style'].value_counts()
```

```python
df['num-of-doors'].value_counts().idxmax()
```

```python
df['num-of-doors'].replace(np.nan, 'four', inplace = True)
```

```python
df.dropna(subset = ['price'], inplace = True, axis = 0)
df.reset_index(drop = True, inplace = True)
```

```python
missing_data = df.isnull()

for columns in headers_data:
  print(columns)
  print(missing_data[columns].value_counts())
  print(" ")

df.head(10)
```

```python
# replace value with scikit learn
import numpy as np
```

```python
from sklearn.impute import SimpleImputer
imputer_mean = SimpleImputer(missing_values = np.nan, strategy = 'mean')
df[['normalized-losses', 'horsepower', 'stroke', 'bore', 'peak-rpm']] = imputer_mean.fit_transform(df[['normalized-losses', 'horsepower', 'stroke', 'bore', 'peak-rpm']]).astype('object')
```

```python
df.head(10)
```

```python
imputer_most_freq = SimpleImputer(missing_values= np.nan, strategy = 'most_frequent')
df[['num-of-doors']] = imputer_most_freq.fit_transform(df[['num-of-doors']])
```

```python
df.dropna(subset=['price'] , inplace=True, axis = 0)
df.reset_index(drop=True, inplace = True)
```

```python
df.info()
```

```python
ls = ['bore', 'stroke', 'peak-rpm', 'price', 'normalized-losses']
df[ls] = df[ls].astype('float')
df.info()
```
